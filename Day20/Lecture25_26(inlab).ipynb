```python
import numpy as np

a = np.random.rand(1000)
b = np.random.rand(1000)

dot_product = 0

for i in range(len(a)):
    dot_product += a[i] * b[i]


print("Dot Product:", dot_product)
```

```python
import numpy as np

a = np.random.rand(1000)
b = np.random.rand(1000)

# Using NumPy's dot product function for efficiency
dot_product = np.dot(a, b)

print("Dot Product:", dot_product)
```
