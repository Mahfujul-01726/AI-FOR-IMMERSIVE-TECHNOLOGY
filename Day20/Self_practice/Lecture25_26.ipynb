# Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
texts = ["I love coding", "Python is amazing", "Coding is fun"]

# TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)

# Display feature names and TF-IDF matrix
print("Feature Names:", vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n", tfidf_matrix.toarray())
```

# Computing on data

```python
import pandas as pd

# Create a sample DataFrame
data = {'height_cm': [170, 165, 180], 'weight_kg': [70, 65, 80]}
df = pd.DataFrame(data)

# Compute BMI: weight (kg) / (height (m))^2
df['height_m'] = df['height_cm'] / 100
df['BMI'] = df['weight_kg'] / (df['height_m'] ** 2)

print(df)
```

# Plotting on data

```python
import matplotlib.pyplot as plt

# Plotting BMI distribution
plt.hist(df['BMI'], bins=5, color='skyblue', edgecolor='black')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot
plt.scatter(df['height_cm'], df['weight_kg'], color='green', marker='o')
plt.title('Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```
